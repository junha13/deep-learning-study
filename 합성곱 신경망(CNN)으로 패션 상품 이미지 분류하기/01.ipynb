```python
import keras
from keras import layers

# 필터 개수와 커널 크기를 지정하는 합성곱층을 추가
# 3,3 짜리 필터를 10개 생성하는거임
conv1 = layers.Conv2D(filters=10, kernel_size=(3,3))
```

```python
import numpy as np

# normal() 함수를 사용해 28, 28, 1 크기의 흑백 이미지 10개를 랜덤으로 생성
x = np.random.normal(size=(10, 28, 28, 1))
# 첫 번째 합성곱층에 x를 입력으로 전달하고 특성 맵을 출력
conv_out = conv1(x)
print(conv_out.shape) # shape는 데이터의 크기 확인 가능

# 특징맵 10개는 26의 크기를 가지고 나옴
# 크기가 28이면 배열로 인덱스값은 27이 나옴. 근데 이전에 정의한 필터의 크기가 3이니 전체를 스캔하려면 시작 위치가 26에 위치해야 27까지 커버 가능
# 27이 시작위치면 28에 해당하는 인덱스가 없기 때문
# 그래서 계속해서 합성곱 연산이 들어가게 되면 모델이 점점 더 작아짐
```

```python
conv2 = layers.Conv2D(filters=10, kernel_size=(3, 3), strides=(2, 2))
print(conv2(x).shape)
```

```python
conv3 = layers.Conv2D(filters=10, kernel_size=(3, 3), strides=(2, 2), padding='same')
print(conv3(x).shape)
```

```python
dense1 = layers.Dense(3)

x2 = np.array([[5, 7]])
print(dense1(x2).shape)
```

```python
print(dense1.get_weights())

dense1.set_weights([
    np.array([[1, 2, 3], [4, 5, 6]]),
    np.array([0, 0, 0])
])

print(dense1(x2))
```
